## Analysis of the MTA turnstile data

I performed exploratory data analysis on the MTA turnstile dataset after cleaning the data (see final cell) and calculating percent change in turnstile traffic for each station between 2015-2019 (data was from March only). My goal is to identify the stations that have a growing number of commuters, indicating potential for residential real estate opportunities.

Stations with the highest percent change in turnstile data between 2015-2019 are presented as a bar plot. Data from all days of the week (left plot) or weekdays (right plot) were used in the calculations. 

These results show that there are stations that when comparing growth in traffic between the all days and weekdays, there are some stations that are within the top ten only when evaluating weekdays. This suggests that these stations are experiencing an increase in traffic from commuters, indicative of a growing residential area. 

I cleaned the MTA turnstile data by doing the following:
* Making each column its appropriate dtype/n 
* Removing any entries that were not 'REGULAR' in the 'DESC' columns/n
* Deleting duplicate data/n
* Updating station names to take account for different uses of abbreviations (e.g. AV and AVE)
* Sorting the data by turnstile and datetime/n
* Find the number of entries and exits from the raw cumulative data by subtracting current entry from previous entry/n
* Replacing Na with 0/n
* Replacing the negative entries and exits (due to the counter counting down instead of up) with the absolute value/n
* Removing outliers by setting a cutoff at the .997 quantile (3 standard deviations above the mean)
    * I will later use a more appropriate cutoff since this data displays a Poisson distribution and is not normally distributed
